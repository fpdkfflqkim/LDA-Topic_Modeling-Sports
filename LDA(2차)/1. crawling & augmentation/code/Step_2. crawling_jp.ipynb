```python
total_title = []
total_date = []
total_journal = []
total_abstract = []
```

```python
for page in range(0,300,20):
    browser = webdriver.Chrome(service=service, options=options)
    browser.get(f"https://www.jstage.jst.go.jp/result/global/-char/ja?fromPage=%2Fsearch%2Fglobal%2F_search%2F-char%2Fen&freeText=&item1=2&word1=%E5%9C%B0%E5%9F%9F%E3%82%B9%E3%83%9D%E3%83%BC%E3%83%84%E3%82%AF%E3%83%A9%E3%83%96&cond1=1&notCond1=0&item2=3&word2=%E5%9C%B0%E5%9F%9F%E3%82%B9%E3%83%9D%E3%83%BC%E3%83%84%E3%82%AF%E3%83%A9%E3%83%96&cond2=2&notCond2=0&item3=&word3=&cond3=&notCond3=&item4=&word4=&notCond4=&count=20&from={page}&order=&type=&license=&attribute=&languageType=&option=&yearfrom=&yearto=&category=&cdjournal=&favorite=&translate=0&bglobalSearch=false&sortby=1&showRecodsH=20&showRecords=20")
    browser.implicitly_wait(10)
    
        # abstract
    all_abstract = browser.find_element(By.XPATH,f"/html/body/form/div/div/div[2]/div/div[2]/div[2]/div[1]/a[2]")
    all_abstract.click()
    time.sleep(1)
    
    # for num in range(1,21):
    #         full = browser.find_element(By.XPATH,f"/html/body/form/div/div/div[2]/div/div[6]/ul/li[{num}]/div[6]/div[2]/a")
    #         full.click()


    for num in range(1,21):
        # 제목
        title = browser.find_element(By.XPATH, f"/html/body/form/div/div/div[2]/div/div[6]/ul/li[{num}]/div[1]/a")
        title_text = title.text
        total_title.append(title_text)
        
        # 저널_이름
        try:
            journal = browser.find_element(By.XPATH, f"/html/body/form/div/div/div[2]/div/div[6]/ul/li[{num}]/div[3]/a")
            journal_text = journal.text
            total_journal.append(journal_text)
        except:
            journal = browser.find_element(By.XPATH, f"/html/body/form/div/div/div[2]/div/div[6]/ul/li[{num}]/div[2]/div/div[2]/a")
            journal_text = journal.text
            total_journal.append(journal_text)
        
        # abstract
        try:          
            full = browser.find_element(By.XPATH,f"/html/body/form/div/div/div[2]/div/div[6]/ul/li[{num}]/div[6]/div[2]/a")
            full.click()
            
            abstract = browser.find_element(By.XPATH,f"/html/body/form/div/div/div[2]/div/div[6]/ul/li[{num}]/div[6]/div[2]")
            text = abstract.text
            
            # total_abstract.append(text)        
        
        except:
            try:          
                full = browser.find_element(By.XPATH,f"/html/body/form/div/div/div[2]/div/div[6]/ul/li[{num}]/div[2]/div/div[5]/div[2]/a")
                full.click()
                
                abstract = browser.find_element(By.XPATH,f"/html/body/form/div/div/div[2]/div/div[6]/ul/li[{num}]/div[2]/div/div[5]/div[2]")
                text = abstract.text
                
                # total_abstract.append(text)        
            
            except:
                text = "None"
        
        total_abstract.append(text)  
```
