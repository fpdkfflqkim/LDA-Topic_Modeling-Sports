```python
from wordcloud import WordCloud, STOPWORDS
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
# all_mask = np.array(Image.open('all.jpg'))
jp_mask = np.array(Image.open(r"C:\Users\user\Desktop\토픽 모델링\3. LDA\jp.jpg"))
kr_mask = np.array(Image.open(r"C:\Users\user\Desktop\토픽 모델링\3. LDA\kr.jpg"))
```

```python
# 빈도 데이터
japan = pd.read_excel(r"C:\Users\user\Desktop\토픽 모델링\3. LDA\data\3.wc용_data_jp.xlsx", index_col=0)
```

```python
# 불용어
sw = list(pd.read_excel(r"C:\Users\user\Desktop\토픽 모델링\3. LDA\data\불용어리스트.xlsx")["stopwords"])

stopwords = set(STOPWORDS)
stopwords.update(sw)
```

```python
text = ' '.join(japan['words'].astype(str))
wordcloud = WordCloud(random_state = 123, width = 800, height = 800, background_color = 'white', mask=jp_mask, stopwords=stopwords, repeat=False, max_words=60, relative_scaling=0).generate(text)
plt.figure(figsize=(20, 20))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
