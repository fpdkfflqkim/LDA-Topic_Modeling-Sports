```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 모든 단어 불러오기
all_tokens = pd.read_excel(r"C:\Users\user\Desktop\토픽모델링_교수님\Data\all_extract_nouns_stopwords.xlsx", index_col=0)
```

```python
text = ' '.join(all_tokens['words'].astype(str))
```

```python
wordcloud = WordCloud(random_state = 123, width = 400, height = 400, background_color = 'white').generate(text)
```

```python
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
