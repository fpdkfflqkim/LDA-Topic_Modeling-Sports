```python
# Author Topic Model 실행 : 있으면 load하고 없으면 model을 만들기
if not os.path.exists(r'C:/Users/user/Desktop/토픽모델링_교수님/Data/ATM/ATM_model_100ep_stopwords'):
    model = AuthorTopicModel(corpus=corpus, num_topics=num_topics, id2word=dictionary.id2token, author2doc=author2doc, passes=100,)
    model.save(r'C:/Users/user/Desktop/토픽모델링_교수님/Data/ATM/ATM_model_100ep_stopwords')

else:
    model = AuthorTopicModel.load(r'C:/Users/user/Desktop/토픽모델링_교수님/Data/ATM/ATM_model_100ep_stopwords')
```

```python
topic_labels = ["Topic0",
                "Topic1",
                "Topic2",
                "Topic3",
                "Topic4",
                "Topic5",
                "Topic6",
                "Topic7",
                "Topic8",
                "Topic9"]
```

```python
# 토픽별로 topN 단어 확인하기.
for topic in model.show_topics(model.num_topics):
    print('Label: ' + topic_labels[topic[0]])
    words = ''
    for word, prob in model.show_topic(topic[0], topn=10):
        words += word + ' '
    print('Words: ' + words)
    print()
```

```python
# 토픽별 단어 가중치(10개)_가중치 포함
top_n = 10

for i in range(num_topics):
    print(f"topic{i}", ":" ,model.get_topic_terms(i, topn=top_n))
```

```python
# 토픽별 단어 가중치(10개)_가중치 미포함
all_corpus_ids = []
for i in range(len(model.get_topics())):
    corpus_ids = [term_id for term_id, _ in model.get_topic_terms(i, topn=top_n)]
    all_corpus_ids.append(corpus_ids)
```

```python
# 저자별 토픽
unique_authors = list(model.author2doc.keys())

for author in unique_authors:
    author_topics = model.get_author_topics(author)
    print(f"Author: {author}")
    
    for topic, weight in author_topics:
        print(f"Topic {topic}: {round(weight, 4)}")
    print("\n")
```

```python
# 퍼플렉서티
doc2author = atmodel.construct_doc2author(model.corpus, model.author2doc)
corpus_words = sum(cnt for document in model.corpus for _, cnt in document)

# Compute bound and divide by number of words.
perwordbound = model.bound(model.corpus, author2doc=model.author2doc, \
                           doc2author=model.doc2author) / corpus_words
print(perwordbound)
```
